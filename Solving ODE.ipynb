(Forward) Euler method to solve first order linear differential equations.

```python
import numpy as np
import matplotlib.pyplot as plt

#set x-grid size and spacing
dx = 0.01
x = np.arange(0, dx+1, dx)
print(x)
N = len(x)
print(N)
#create y_grid
y = np.zeros(N)
print (y, len(y))
y[0]=2
print(y)
for i in range(0,N-1):
    #the Euler method
    y[i+1]= y[i] + dx * (x[i] * y[i])
print(y)

plt.plot(x,y)
```
