(Forward) Euler method to solve first order linear differential equations.

```python
import numpy as np
import matplotlib.pyplot as plt

#set x-grid size and spacing
dx = 0.01
x = np.arange(0, dx+1, dx)
print(x)
N = len(x)
print(N)
#create y_grid
y = np.zeros(N)
print (y, len(y))
y[0]=2
print(y)
for i in range(0,N-1):
    #the Euler method
    y[i+1]= y[i] + dx * (x[i] * y[i])
print(y)

plt.plot(x,y)
```

```python
import numpy as np
from scipy.integrate import solve_ivp

#plot exact solution first
time = np.linspace(0,3)
y_exact = ( -6 * np.exp(-3*time) + 7*np.exp(-2*time) + np.sin(time) - np.cos(time))
plt.plot(time, y_exact, label='Exact')

def mass_spring(time, z):
    #Function providing dz/dt derivative system
    omega = 1
    return [z[1], 10*np.sin(omega*time) - 6*z[0] - 5*z[1]]

#first argumenti is the function
#then we have the range
#then we have the initial values
#then we have the method we want to use
sol = solve_ivp(mass_spring, [0, 3], [0, 5], method='RK45')

#the soulition has two columns, we are interested in the first
plt.plot(
    sol.t, sol.y[0,:], '.',
    markersize=15, label='RK45 (solve_ivp)'
)
plt.grid(True)
plt.legend()
plt.show()
```
